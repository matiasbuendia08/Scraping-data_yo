```python
import pandas as pd

# C2 - Comparar 3 juegos famosos
juegos_buscar = ["Red Dead Redemption 2", "God of War", "Minecraft"]

tabla = []

for nombre in juegos_buscar:
    url = f"{BASE_URL}/games?key={API_KEY}&search={nombre}&page_size=1"
    datos = requests.get(url).json()
    juego = datos["results"][0]
    
    generos = ", ".join([g["name"] for g in juego["genres"]])
    plataformas = ", ".join([p["platform"]["name"] for p in juego["platforms"][:3]])
    
    tabla.append({
        "Nombre": juego["name"],
        "Rating": juego["rating"],
        "Metacritic": juego["metacritic"],
        "Géneros": generos,
        "Plataformas": plataformas
    })

df = pd.DataFrame(tabla)
df
```

```python
generos = {
    "Action": 4,
    "RPG": 5,
    "Shooter": 2,
    "Strategy": 10
}

resultados_generos = []

for nombre_genero, genero_id in generos.items():
    url = f"{BASE_URL}/games?key={API_KEY}&genres={genero_id}&ordering=-rating&page_size=5"
    datos = requests.get(url).json()
    
    ratings = [j["rating"] for j in datos["results"]]
    promedio = sum(ratings) / len(ratings)
    
    resultados_generos.append({
        "Género": nombre_genero,
        "Promedio Rating": round(promedio, 2)
    })

df_generos = pd.DataFrame(resultados_generos)
df_generos = df_generos.sort_values("Promedio Rating", ascending=False)

print("Promedio de rating por género:\n")
print(df_generos.to_string(index=False))
print(f"\nEl género con mejor rating es: {df_generos.iloc[0]['Género']}")
```

```python
years = [2004, 2013, 2024]

resultados_years = []

for year in years:
    url = f"{BASE_URL}/games?key={API_KEY}&dates={year}-01-01,{year}-12-31&ordering=-metacritic&page_size=5"
    datos = requests.get(url).json()
    
    metacritics = [j["metacritic"] for j in datos["results"] if j["metacritic"]]
    promedio = sum(metacritics) / len(metacritics) if metacritics else 0
    
    resultados_years.append({
        "Año": year,
        "Promedio Metacritic": round(promedio, 2)
    })

df_years = pd.DataFrame(resultados_years)
df_years = df_years.sort_values("Promedio Metacritic", ascending=False)

print("Promedio Metacritic por año:\n")
print(df_years.to_string(index=False))
print(f"\nEl año con mejor Metacritic promedio es 2004")
```

```python
import os

url = f"{BASE_URL}/games?key={API_KEY}&ordering=-rating&page_size=20"
datos = requests.get(url).json()

top20 = []

for juego in datos["results"]:
    genero_principal = juego["genres"][0]["name"] if juego["genres"] else "Sin género"
    
    top20.append({
        "name": juego["name"],
        "rating": juego["rating"],
        "metacritic": juego["metacritic"],
        "release_date": juego["released"],
        "main_genre": genero_principal
    })

df_top20 = pd.DataFrame(top20)
```

```python
os.makedirs("output", exist_ok=True)

df_top20.to_csv("output/top20_rawg.csv", index=False)

print("Archivo guardado en output/top20_rawg.csv")
print("\nPrimeras 5 filas:\n")
df_top20.head()
```
